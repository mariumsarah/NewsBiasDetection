# This Python file is used to extract articles their text and headlines
## Specifically from the following news outlets due to their large number of Democrate/Republican Bias articles
### DailyKos and Breitbard

```python
import pandas as pd
from urllib import request
from bs4 import BeautifulSoup
```

```python
articles_set = pd.read_csv("articles.csv")
```

```python
articles_set['headline'] = ""
articles_set['content'] = ""
articles_set
```

## Step 1: Add Headlines into the Dataframe for each article

```python
# Add Headline 
for index, row in articles_set.iterrows():
    print(index)
    url_1 = row["url"]
    try:
        page = request.urlopen(url_1)
    except request.HTTPError:
        articles_set.drop(articles_set.index[index])
        continue
    soup = BeautifulSoup(page)
    title = soup.find('title')
    row["headline"]=title.string
```

## Step 2: Add Content into the Dataframe for each article

```python
articles_set['content'] = ""
for index, row in articles_set.iterrows():
    print(index)
    url_1 = row["url"]
    try:
        page = request.urlopen(url_1)
    except request.HTTPError:
        continue
    soup = BeautifulSoup(page)
    title = soup.find('title')
    print(title.string)
    if(row['News Outlet']=="Breitbart"):
        article_text = ''
        article = soup.find("div", {"class" : "entry-content"}).findAll(['p','h2'])
        for element in article:
            article_text += '\n' + ''.join(element.findAll(text = True))
        row["content"]=article_text
    else:
        article_text = ''
        articles = soup.findAll("div", {"class" : "story-content"})
        for n in articles:
            article_text += '\n'+ n.get_text()
        row["content"]=article_text
    
```
